```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head(20) #for inspecting tables and columns top few rows i used df.head()
```

```python
df.info() #I use df.info() for getting basic overview of the structure of table
# We found that there are 3900 rows and 18 columns with three datatype available (float64(1), int64(4), object(13))
# There are few null values in Review ratings
```

```python
df.describe(include = 'all') #It is used for finding statistical summary of entire table
```

```python
df.isnull().sum() # Found 37 Null values in review Rating
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.info()
```

```python
# Creating a new column
group_list = ['Young Adult', 'Adult', 'Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = group_list)
```

```python
df[['age','age_group']].head()
```

```python
df['age_group'].value_counts()
```

```python
# create column purchase frequency days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly' : 90,
    'Bi-Weekly': 14,
    'Annually' : 365,
    'Every 3 Months' : 90}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop(['promo_code_used'],axis = 1)
df.head()
```
